```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import torch
import torch.nn as nn
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from PIL import Image
import torchvision.transforms as transforms

imsize = 256
img_transforms = transforms.Compose([
    transforms.Resize((imsize, imsize)),
    transforms.ToTensor()
])
```

```python
def image_loader(image_name):
    image = Image.open(image_name)
    image = img_transforms(image).unsqueeze(0)
    return image.to(device, torch.float)

style_img = image_loader("/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/style.jpg")
content_img = image_loader("/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/content.jpg")
```

```python
def imshow(tensor, title=None):
    image = tensor.cpu().clone()
    image = image.squeeze(0)
    image = transforms.ToPILImage()(image)
    plt.imshow(image)
    if title is not None:
        plt.title(title)


plt.figure()
imshow(style_img, title='Style Image')

plt.figure()
imshow(content_img, title='Content Image')
```

# Loss funtion

```python
content_weight = 1
ContentLoss = nn.MSELoss()
```

Style loss

```python
def gram_matrix(tensor):
    a, b, c, d = tensor.size()
    tensor = tensor.view(a * b, c * d)
    G = torch.mm(tensor, tensor.t())
    return G.div(a * b * c * d)

style_weight = 1e6
StyleLoss = nn.MSELoss()
```

# Model

```python
from torchvision.models import vgg19, VGG19_Weights

VGG19_pretrained = vgg19(weights=VGG19_Weights.DEFAULT).features.eval()
for param in VGG19_pretrained.parameters():
    param.requires_grad_(False)


VGG19_pretrained.to(device)
```

```python
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']
```

```python
def get_features(pretrained_model, image):
    layers = {
        '0': 'conv_1',
        '5': 'conv_2',
        '10': 'conv_3',
        '19': 'conv_4',
        '28': 'conv_5'
    }
    features = {}
    x = image
    for name, pretrained_layer in pretrained_model._modules.items():
        x = pretrained_layer(x)
        if name in layers:
            features[layers[name]] = x
    return features
```

# Training

```python
content_features = get_features(VGG19_pretrained, content_img)
style_features = get_features(VGG19_pretrained, style_img)
```

```python
import torch.optim as optim

target_img = content_img.clone().requires_grad_(True).to(device)
optimizer = optim.Adam([target_img], lr=0.02)
```

```python
STEPS = 2000

for step in range(STEPS):
    optimizer.zero_grad()
    target_features = get_features(VGG19_pretrained, target_img)
    content_loss = ContentLoss(content_features['conv_1'], target_features['conv_1'])

    style_loss = 0
    for layer in style_layers:
        target_gram = gram_matrix(target_features[layer])
        style_gram = gram_matrix(style_features[layer])
        style_loss += StyleLoss(style_gram, target_gram)

    total_loss = content_loss*content_weight + style_loss*style_weight
    total_loss.backward()
    optimizer.step()

    with torch.no_grad():
        target_img.clamp_(0, 1)

    if step % 100 == 0:
        print(f"Epoch [{step+1}/{STEPS}] total loss: {total_loss.item():.8f}")

        fname = '/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/at_%d.png' % (step+1)
        print(target_img[0].detach().cpu().permute(1, 2, 0).numpy().shape)
        mpl.image.imsave(fname, target_img[0].detach().cpu().permute(1, 2, 0).numpy())
```

```python
plt.figure()
imshow(target_img.detach(), title='Output Image')
plt.show()
```

# Dual Style Transfer

```python
def get_dual_style(style_features1, style_features2, style_layers):
    final_style_features = {}
    for layer in style_layers:
        sf1 = style_features1[layer]
        sf2 = style_features2[layer]
        sf1_size = int(sf1.size()[1] / 4)
        final_style_features[layer] = torch.concatenate([sf1[: , :sf1_size ,: ,:] ,
                                                         sf2 [: , sf1_size: ,: ,:]] , dim =1)
    return final_style_features
```

```python
style_img = image_loader("/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/style.jpg")
content_img = image_loader("/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/content.jpg")
style2_img = image_loader("/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/style2.jpg")
```

```python

plt.figure()
imshow(style_img, title='Style Image')

plt.figure()
imshow(content_img, title='Content Image')


plt.figure()
imshow(style2_img, title='Style2 Image')

```

```python
content_features = get_features(VGG19_pretrained, content_img)
style_features1 = get_features(VGG19_pretrained, style_img)
style_features2 = get_features(VGG19_pretrained, style2_img)
final_style_features = get_dual_style(style_features1, style_features2, style_layers)
```

```python
target_img = content_img.clone().requires_grad_(True).to(device)
optimizer = optim.Adam([target_img], lr=0.02)
```

```python
STEPS = 2000

for step in range(STEPS):
    optimizer.zero_grad()
    target_features = get_features(VGG19_pretrained, target_img)
    content_loss = ContentLoss(content_features['conv_1'], target_features['conv_1'])

    style_loss = 0
    for layer in style_layers:
        target_gram = gram_matrix(target_features[layer])
        style_gram = gram_matrix(final_style_features[layer])
        style_loss += StyleLoss(style_gram, target_gram)

    total_loss = content_loss*content_weight + style_loss*style_weight
    total_loss.backward()
    optimizer.step()

    with torch.no_grad():
        target_img.clamp_(0, 1)

    if step % 100 == 0:
        print(f"Epoch [{step+1}/{STEPS}] total loss: {total_loss.item():.8f}")

        fname = '/content/drive/MyDrive/Module_9_CV_Advance/Style transfer/image result/at_%d.png' % (step+1)
        print(target_img[0].detach().cpu().permute(1, 2, 0).numpy().shape)
        mpl.image.imsave(fname, target_img[0].detach().cpu().permute(1, 2, 0).numpy())
```
